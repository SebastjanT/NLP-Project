# Basic text analysis of the 7 provided short stories from the IMapBook project

```python
import sys
sys.path.append(".")

import os
import string
import nltk
import matplotlib

from pathlib import PurePath
```

```python
# Prepare text
def getTokens(name, special_chars):
   with open(name, 'r') as shakes:
    text = shakes.read().lower()
    # remove punctuation
    table = text.maketrans({key: None for key in string.punctuation + special_chars})
    text = text.translate(table)      
    tokens = nltk.word_tokenize(text)
    return tokens
```

```python
%matplotlib inline
# Process the text files
data_path = PurePath("../data/imapbook")

stopwords = nltk.corpus.stopwords.words("english")
special_chars = "“”,’"

for path, _, files in os.walk(data_path):
    for name in files:
        print(f"Basic analysis of {name}:")
        # Retrieve text and tokens
        tokens = getTokens(PurePath(data_path, name), special_chars)
        print(f"\nNumber of tokens in text: {len(tokens):.>50}")
        # Remove the stop-words
        tokens_stripped = [token for token in tokens if token not in stopwords]
        print(f"\nNumber of tokens in text after stop-words removal: {len(tokens_stripped):.>50}")
        # Get the frequency distribution of the tokens
        print("\nPlotting 10 most frequent tokens in text:")
        tokens_stripped_dist = nltk.FreqDist(tokens_stripped)
        tokens_stripped_dist.plot(10)
        # Perform common context on most frequent word
        most_frequent = tokens_stripped_dist.max()
        text = nltk.Text(tokens)
        print(f"\nCommon contexts to the most frequent word '{most_frequent}':")
        text.common_contexts([most_frequent])
        print(f"\nContexts of the most frequent word '{most_frequent}':")
        text.concordance(most_frequent)
        print("\n\n\n")
        
```

## Observations

* Files `Henry_Red_Chief.txt`, `the_gift_of_the_magi_0_Henry.txt` and `Hills Like White Elephants.txt` had to be reincoded in utf-8 as for sucessfull parsing (previous encoding `Western (Mac Roman)`)
* Extra punctuation characters had to be added to be removed because of the above files
